```python
# Age distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()

# Fare distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
# Boxplot of Age vs Survived
plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age vs Survival')
plt.show()

# Boxplot of Fare vs Survived
plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare vs Survival')
plt.show()
```

```python
# Scatterplot: Age vs Fare
plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Age vs Fare (colored by Survival)')
plt.show()

# Scatterplot: Fare vs Pclass
plt.figure(figsize=(8,5))
sns.scatterplot(x='Pclass', y='Fare', hue='Survived', data=df)
plt.title('Pclass vs Fare (colored by Survival)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecting only relevant continuous columns
cols = ['Age', 'Fare', 'SibSp', 'Parch']

# Pairplot for the continuous columns with hue based on 'Survived' column
sns.pairplot(df[cols + ['Survived']], hue='Survived', palette='coolwarm', markers=['o', 's'])
plt.suptitle('Pairplot - Titanic Dataset', y=1.02, fontsize=16)
plt.show()
```

```python
# Calculate the correlation matrix
corr = df[['Age', 'Fare', 'SibSp', 'Parch', 'Survived']].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)
plt.title('Correlation Heatmap - Titanic Dataset', fontsize=16)
plt.show()
```
