```python
df['Education'] = df['Education'].str.strip().str.lower()
df['Marital_Status'] = df['Marital_Status'].str.strip().str.lower()
df.head()
```

#### we have standardize the columns like "education" and "Martial_Status" because these are the categorical columns so we make all data into standardize lower case form . Remaining don't need beacause all are numerical columns

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y', errors='coerce')
df.head()

### errors='coerce': If pandas encounters an invalid date or a value that cannot be converted to a valid datetime, it will replace that value with NaT (Not a Time), which is similar to NaN for datetime values. This ensures that the conversion process continues without raising an error.
```

```python
print(df['Dt_Customer'].head(10))
## pandas usually displays datetime in yyyy-mm-dd format only
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.head()
```

```python
#### .str.strip() → removes leading/trailing spaces

#### .str.lower() → makes everything lowercase

#### .str.replace(' ', '_') → replaces spaces with underscores

#### Uniform formatting (lowercase, underscores) makes code readable and consistent
```
